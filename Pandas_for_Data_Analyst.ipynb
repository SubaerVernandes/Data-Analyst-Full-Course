The best way to think about the pandas data structures is as flexble containers for lower dimensional data. For example, DataFrame is a container for SEries, and Series is a container for scalars. ##### Series in pandas:

Pandas series is a one-dimensional labeled array capable of holding data of any type (integer, string float, python objects, etc) The axis labels are collectively called indes. Pandas series is nothing but a column in an excel sheet.

The object supports both integer and label-based indexing and provides a host of methods for performing operations involving the index.


##### DataFrames:

Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is two-dimensional data structure. i.3., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

##### merge, join and concatenate in pandas

```python
data1 = {"Emp Id"   : ["EO1", "E02", "E03", "E04", "E05", "E06"],
        "Names"     : ["Ram", "Shym", "Rahul", "Vishal", "Ravi", "John"],
        "Age"       : [34, 56, 23, 44, 32, 36] }

data2 = {"Emp Id"   : ["E01", "E02", "E03", "E04", "E05", "E06"],
         "Salary"    :[45000, 56000, 34000, 30000, 50000, 62000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
# merge (gabungkan)
print (pd.merge(df1, df2, on = "Emp Id"))
```

```python
data3 = {"Emp Id"   : ["E01", "E02", "E03", "E04", "E05", "E06"],
        "Names"     : ["Ram", "Shym", "Rahul", "Vishal", "Ravi", "John"],
        "Age"       : [34, 56, 23, 44, 32, 36] }

data4 = {"Emp Id"   : ["E01", "E07", "E03", "E04", "E08", "E06"],
         "Salary"    :[45000, 56000, 34000, 30000, 50000, 62000]}

df3 = pd.DataFrame(data3)
df4 = pd.DataFrame(data4)

print(df3)
print()
print(df4)
```

```python
# merge (gabungkan)
print(pd.merge(df3, df4, on = "Emp Id")) # digabungkan berdasarkan nilai yang sama pada kolom "Emp Id" (seperti subset di matematika)
```

```python
# semua data dari DataFrame kiri (df1) akan tetap ditampilkan,
# lalu pandas akan mencari data yang cocok di DataFrame kanan (df2),
# jika tidak ditemukan pasangan, maka nilainya akan diisi NaN.

# Metode ini berguna ketika kita ingin mempertahankan seluruh data utama pada tabel kiri sambil menambahkan informasi dari tabel kanan jika tersedia.

print(pd.merge(left=df3, right=df4, on="Emp Id", how="left"))
```

```python
# Gabungkan df1 dan df2 berdasarkan "Emp Id",
# tampilkan semua data dari DataFrame kanan (df2),
# lalu ambil data yang cocok dari DataFrame kiri (df1).

# Jika tidak ada pasangan di df1, maka nilainya akan menjadi NaN.

print(pd.merge(left=df3, right=df4, on="Emp Id", how="right"))
```

##### pandas | compare DataFrame

```python
import pandas as pd 

dict = {"Fruits"    : ["mango", "apples", "banana", "papaya"],
        "Price"     : [100, 150, 50, 35],
        "Quantity"  : [15, 10, 10, 3]}

df7 = pd.DataFrame(dict)
print(df7)
```

```python
df8 = df7.copy()
print(df8)
```

```python
# mengubah data pada posisi tertentu dalam DataFrame.
# mengedit nilai sesuai index baris dan kolomnya
# .loc dipakai untuk mengakses atau mengubah data berdasarkan: baris dan nama kolom
# DataFrame.loc[index_baris, "nama_kolom"] = nilai_baru
df8 = df7.copy()

df8.loc[0, "Price"] = 120
df8.loc[1, "Price"] = 175
df8.loc[3, "Price"] = 30
df8.loc[0, "Quantity"] = 12
df8.loc[1, "Quantity"] = 15
df8.loc[3, "Quantity"] = 5

print(df8)

# compare()
# melihat perubahan data dan hanya menampilkan nilai yang berubah (self --> awal other --> perubahan)
print()
print(df7.compare(df8))
```

```python
# untuk melihat perubahan data pada banyak kolom
print(df7.compare(df8, align_axis = 0))
```

```python
# melihat seluruh isi baris yang berubah dan membandingkan data lama vs baru secara lengkap
print(df7.compare(df8, keep_equal = True))
```

```python
# menampilkan semua data walaupun tidak berubah, yg tidak berubah ditampilkan sebagai NaN
print(df7.compare(df8, keep_shape = True))
```

```python
# melihat perubahan data saja yg tidak berubah tidak di tampilkan
print(df7.compare(df8, keep_shape = False))
```

##### pivoting and melting dataframes

```python
dict = {"keys"  : ["k1", "k2", "k1", "k2"],
        "Names" : ["John", "Ben", "David", "Peter"],
        "Houses": ["red", "blue", "green", "red"],
        "Grades": ["3rd", "8th", "9th", "8th"]}

df9 = pd.DataFrame(dict)
print(df9)

# mirip seperti pivot table di Excel
# keys dijadikan baris, Names dijadikan kolom, isi tabel diambil dari Houses dan Grades
print()
print(df9.pivot(index = "keys", columns = "Names", values = ["Houses", "Grades"]))
```

```python
dict = {
        "Names" : ["John", "Ben", "David", "Peter"],
        "Houses": ["red", "blue", "green", "red"],
        "Grades": ["3rd", "8th", "9th", "8th"]}
df10 = pd.DataFrame(dict)
print(df10)

# melting
# digunakan untuk data preprocessing, visualisasi data, machine learning, statistik, mengubah data ke format long format
# cara baca ouput artinya: John data berasal dari kolom Houses nilainya red
print()
print(pd.melt(df10, id_vars = ["Names"], value_vars = ["Houses", "Grades"]))
```

```python
# melting / melt
# nama kolom lebih jelas, mudah dipahami, sesuai kebutuhan laporan/analisis
print(pd.melt(df10, id_vars = ["Names"], value_vars = ["Houses", "Grades"], var_name = "Houses&Grades", value_name="values"))
```
