##### Python Aritmatic Operator

```python
# Nilai contoh
a = 10
b = 3

# Addition (Penjumlahan)
addition = a + b
print("Addition:", addition)  # 13

# Subtraction (Pengurangan)
subtraction = a - b
print("Subtraction:", subtraction)  # 7

# Multiplication (Perkalian)
multiplication = a * b
print("Multiplication:", multiplication)  # 30

# Division (Pembagian)
division = a / b
print("Division:", division)  # 3.3333...

# Floor Division (Pembagian bulat ke bawah)
floor_division = a // b
print("Floor Division:", floor_division)  # 3

# Modulus (Sisa bagi)
modulus = a % b
print("Modulus:", modulus)  # 1

# Exponentiation (Pangkat)
exponentiation = a ** b
print("Exponentiation:", exponentiation)  # 1000
```

#####  Python Comparison Opertators

```python
# Nilai contoh
a = 10
b = 5

# Equal to (==)
print("a == b:", a == b)  # False

# Not equal to (!=)
print("a != b:", a != b)  # True

# Greater than (>)
print("a > b:", a > b)  # True

# Less than (<)
print("a < b:", a < b)  # False

# Greater than or equal to (>=)
print("a >= b:", a >= b)  # True

# Less than or equal to (<=)
print("a <= b:", a <= b)  # False
```
