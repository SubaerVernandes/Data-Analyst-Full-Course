##### data visualization

Data visualization is the graphical representation of information and data 

in the world of Big Data, data visualization tools and technologies are essential to analyze massive amounts of information and make data-driven decisions.

##### matplotlib

matplotlib is low-level graph plotting library in python that serves as visualization utility

matplotlib was created by John D.Hunter

matplotlib is open-source and we can use it freely.

##### Bar plot matplotlib

```python
import matplotlib.pyplot as plt

y = [98, 67, 88, 95, 88]
x = ["Part1", "Part2", "Part3", "Part4", "Part5"]

plt.bar(x, y)
plt.show()
```

```python
plt.bar(x, y)
plt.xlabel("Parts of Harry Potter")
plt.ylabel("Popularity")
plt.title("Popularity of Different parts of Harry potter")
plt.show()
```

```python
colors = ["red", "green", "yellow", "blue", "orange"]
plt.bar(x, y, color=colors, edgecolor = "black")
plt.xlabel("Parts of Harry Potter", fontsize = 17)
plt.ylabel("Popularity", fontsize = 17)
plt.title("Popularity of Different parts of Harry Potter", fontsize = 20)
plt.show()
```

##### Pie chart matplotlib

```python
import matplotlib.pyplot as plt

brands = ["Oneplus", "Apple", "Samsung", "Nokia", "Redmi"]
x = [22, 35, 30, 3, 10]

plt.pie(x, labels = brands)
plt.show()
```

```python
c = ["red", "violet", "blue", "magenta", "orange"]
ex = [0, 0, 0, 0, 0.1]
plt.pie(x, labels = brands, colors = c , explode = ex, shadow = True, autopct = "%.2f", startangle = 180)
```

```python
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)


grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
plt.pie(grouped_by.values, labels = grouped_by.index, autopct = "%.2f")
plt.show()
```

##### Stack plot matplotlib

```python
days = [1, 2, 3, 4, 5, 6, 7]

NOP1 = [5, 10, 30, 20, 35, 60, 80]
NOP2 = [50, 60, 30, 75, 80, 90, 120]
NOP3 = [8, 30, 50, 50, 70, 90, 100]

plt.stackplot(days, NOP1, NOP2, NOP3, colors=["pink", "hotpink", "magenta"], labels = ["week1", "week2", "week3"])
plt.legend()
plt.show()
```

##### Step plot matplotlib

```python
x = ["day1", "day2", "day3", "day4", "day5"]
y = [30, 40, 25, 30, 40]

plt.step(x, y, where = "mid", marker = "o")
plt.show()
```

```python
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)

group = df.groupby("Category").agg({"Amount" : "sum"})
print(group)

plt.step(group.index, group["Amount"], where = "mid", marker = "0")
plt.show()
```

##### Legend matplotlib
