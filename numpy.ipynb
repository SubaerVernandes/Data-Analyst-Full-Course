##### mathematical operations and functions on arrays

```python
arr1 = np.array ( [ [30, 40], [20, 10] ])
arr2 = np.array ([ [30, 20], [10, 20] ])
print (arr1 + arr2)
print (arr1 - arr2)
print (np.add (arr1, arr2))
print (np.subtract (arr1, arr2))

print ("\n")
print (arr1 * arr2)
print (np.multiply (arr1, arr2))

print ("\n")
print (arr1 / arr2)
print (np.divide (arr1, arr2))
```

```python
arr1 = np.array ([3, 4, 2, 1])
arr2 = np.array ( [3] )
print (np.power (arr1, arr2)) # setiap elemen arr1 di pangkatkan elemen arr2
```

```python
arr1 = np.array ( [9, 16, 4, 1] )
print (np.sqrt (arr1) ) # akar kuadrat dari elemen arr1
```

##### Numpy - sort, filter and search

```python
ar = np.array ( [3, 4, 1, 7, 8])
s = np.where(ar == 4) # search index the elemen
s2 = np.where(ar % 2 == 0)
print (s)
print (s2)
```

```python
ar = np.array ( [1, 2, 3, 4, 5] )
ss = np.searchsorted (ar, 5) # search index the elemen
print (ss)
```

##### aggregating functions in numpy

```python
a = np.array( [20, 40, 60, 70] )

print (np.sum(a))
print (np.min(a))
print (np.max(a))
print (np.size(a))
print (np.mean(a))
print (np.cumsum(a))
print (np.cumprod(a))
```

```python
a = [100, 150, 199, 200, 250, 130]
b = [10, 50, 30, 40, 30, 10]
price = np.array(a)
quantity = np.array(b)

print (price, "\n", quantity)
print ()
print (np.cumprod([price, quantity], axis = 0))

# can to
print ("\n")
c = np.cumprod( [price, quantity], axis = 0)
print (c[1].sum())
```

##### statistical functions numpy

```python
import statistics as stats
baked_food = [200, 150, 150, 130, 200, 220, 170, 188]

a = np.array (baked_food)

print (np.mean(baked_food))     # sum of all the values/number of values
print (np.median(baked_food))   # central value after sorting
print (stats.mode(baked_food))
print (np.std(baked_food))
print (np.var(baked_food))
```

```python
# - 1 represent iversely proportional relationship
# 1 represents proportional relationship
# 0 means no relationship
tobacco_consumption = [30, 50, 10, 30, 50, 40]
deaths = [100, 120, 70, 100, 120, 112]

print (np.corrcoef( [tobacco_consumption, deaths]))
```

```python
price = [300, 100, 350, 150, 200]
sales = [10, 20, 7, 17, 3]

print (np.corrcoef([price, sales]))
```
