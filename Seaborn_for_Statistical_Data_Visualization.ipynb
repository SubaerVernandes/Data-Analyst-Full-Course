##### Data Visualization 

Data visualization is the graphical represenatation of information and data

in the world of big Data, data visualization tools and technologies are essential to analyze massive amounts of information and make data-driven decision.

seaborn is a python data visualization library based on matplotlib it provides high-level interface for drawing atractive and informative statistical graphics.

##### LIne plot seaborn

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt

data = {"Days" : [1, 2, 3, 4, 5],
        "NOP" : [50, 40, 60, 30, 44]}
        
df = pd.DataFrame(data)
print(df)

sns.lineplot(data = data, x = "Days", y = "NOP")
plt.show()
```

```python
import seaborn as sns
data = pd.read_excel("ESD.xlsx")

color = sns.color_palette("GnBu")
sns.lineplot(data = data, x = "Business Unit", y = "Age", hue = "Ethnicity", palette = color)
plt.show()
```

```python
data = sns.load_dataset("tips")
print(data.head(3))
sns.barplot(data = data, x = "day", y = "tip")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "spring", order = ["Sun", "Sat", "Fri", "Thur"])
plt.show()
```

##### Hist plot seaborn

```python
sns.histplot(data, x = "day", hue = "sex", kde = True)
plt.show()
```

```python
data = sns.load_dataset("titanic")
sns.histplot(data = data, x = "age", hue = "who", kde = True)
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who", kde = True, bins = 30)
plt.show()
```

##### scatter plot seaborn

```python
data = sns.load_dataset("tips")

sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size", marker = "o")
```

```python
data = pd.read_excel("ESD.xlsx")
print(data.head(3))

sns.scatterplot(data = data, x = "Age", y = "Annual Salary", size = "Bonus %", hue = "Department")
plt.legend(bbox_to_anchor = (0.2, 0, 1.2, 1)) # x, y, width, height
plt.show()
```

##### Heatmap Seaborn

```python
data = sns.load_dataset("tips")
print(data)

gp = data.groupby("day").agg({"tip" : "mean"})
sns.heatmap(gp)
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
gp = data.groupby(["Job Title"]).agg({"Annual Salary" : "mean"})
sns.heatmap(gp)
plt.show()
```

##### Style and color in plots

```python
data = sns.load_dataset("exercise")

sns.set_style(style = "darkgrid")
sns.barplot(x = "time", y = "pulse", data = data)
plt.show()
```

```python
sns.set_style(style = "whitegrid")
sns.barplot(x = "time", y = "pulse", data = data)
plt.show()
```

```python
sns.set_style(style = "ticks")
sns.barplot(x = "time", y = "pulse", data = data)
plt.show()
```

```python
sns.palplot(sns.color_palette())
plt.show()
```

```python
sns.palplot(sns.color_palette("viridis"))
plt.show()
```

```python
sns.palplot(sns.color_palette("spring"))
plt.show()
```

##### Multiple plots in seaborn

```python
data = sns.load_dataset("tips")
print(data.head(3))

a = sns.FacetGrid(data, col = "smoker", height = 2)
a.map(sns.barplot, "day", "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "smoker", height = 2)
a.map(sns.scatterplot, "day", "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "time", height = 2, hue = "sex")
a.map(sns.scatterplot, "day", "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "time", height = 2, hue = "sex")
a.map(sns.barplot, "day", "tip")
plt.show()
```
